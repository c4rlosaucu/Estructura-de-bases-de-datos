#Problema 1

```python
import hashlib
import itertools

def encontrar_secuencia(hash_objetivo):
    print(f"Buscando secuencia...\n")
    digitos = '0123456789'
    contador = 0

    for combinacion in itertools.product(digitos, repeat=10):
        secuencia = ''.join(combinacion)
        hash_generado = hashlib.sha256(secuencia.encode()).hexdigest()
        contador += 1

        if contador % 1_000_000 == 0:
            print(f"Probadas: {contador:,} combinaciones, última: {secuencia}")

        if hash_generado == hash_objetivo:
            print(f"\n¡Secuencia encontrada!: {secuencia}")
            print(f"Intentos realizados: {contador:,}")
            return secuencia

    print("No se encontró ninguna secuencia.")
    return None

hash_rapido = "90000f64b9eb64d89066d8ec31b960a8bc57cb197c7602fb750e3358cb6d8f20"
encontrar_secuencia(hash_rapido)
```

#Problema 2

```python
import hashlib
from itertools import permutations

def hash_sha256(data):
    return hashlib.sha256(data.encode()).hexdigest()

def merkle_root(transacciones):
    h1 = hash_sha256(transacciones[0])
    h2 = hash_sha256(transacciones[1])
    h3 = hash_sha256(transacciones[2])
    h4 = hash_sha256(transacciones[3])
    h12 = hash_sha256(h1 + h2)
    h34 = hash_sha256(h3 + h4)
    return hash_sha256(h12 + h34)

def encontrar_orden(transacciones, hash_objetivo):
    print(f"Transacciones: {transacciones}")
    print(f"Hash objetivo: {hash_objetivo}\n")
    print("Probando las 24 combinaciones posibles...\n")

    for i, orden in enumerate(permutations(transacciones)):
        raiz = merkle_root(list(orden))
        print(f"  [{i+1:2}] {list(orden)} → {raiz[:20]}...")
        if raiz == hash_objetivo:
            print(f"\n¡Orden encontrado en el intento #{i+1}!")
            print(f"Orden correcto: {list(orden)}")
            print(f"Hash raíz: {raiz}")
            return list(orden)

    print("No se encontró ningún orden.")
    return None


transacciones = ["TX_A", "TX_B", "TX_C", "TX_D"]
orden_secreto  = ["TX_D", "TX_B", "TX_A", "TX_C"]   # esto no lo sabe el programa
hash_objetivo  = merkle_root(orden_secreto)

encontrar_orden(transacciones, hash_objetivo)
```
